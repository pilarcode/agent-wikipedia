```python
from agentwiki.retriever_agent_mem import agent_executor
from agentwiki.output_parser import extract_between_tags
from pprint import pprint

```

# Test Case 1

```python
chat_history = []
query = "Which city has  more citizens: Alicante or Murcia?"
```

```python
response  = agent_executor.invoke({"query": query, "chat_history": chat_history})
```

```python
response
```

```python
response.keys()
```

```python
pprint(response['output'])
```

```python
print(extract_between_tags('information', response['output']))
```

```python
result = extract_between_tags('information', response['output'])
result
```

# Adding memory

```python
from langchain_core.messages import AIMessage, HumanMessage

chat_history.extend(
    [
        HumanMessage(content=query),
        AIMessage(content=result),
    ]
)
```

```python
chat_history
```
